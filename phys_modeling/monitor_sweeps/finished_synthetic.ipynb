```python
"""Imports."""

from pathlib import Path
```

```python
"""Check whether all evaluation stats files are present."""

BASE_RESULTS_DIR = Path("../../cache/modeling/synthetic")
SESSIONS = [
    ("Elgar", "2022-08-19"),
    ("Elgar", "2022-08-20"),
    ("Elgar", "2022-08-22"),
    ("Elgar", "2022-08-24"),
    ("Elgar", "2022-09-04"),
    ("Perle", "2022-05-30"),
    ("Perle", "2022-05-31"),
    ("Perle", "2022-06-01"),
    ("Perle", "2022-06-03"),
    ("Perle", "2022-06-04"),
]
SEEDS = [3, 4, 5, 6, 7]
DATASETS = [
    "triangle_gain",
    "triangle_switching",
    "triangle_slot_partition",
]
MODELS = [
    "triangle_gain",
    "triangle_switching",
    "triangle_slot_partition",
]

for dataset in DATASETS:
    for model in MODELS:
        for subject, session in SESSIONS:
            for seed in SEEDS:
                seed_dir = BASE_RESULTS_DIR / dataset / model / subject / session / str(seed)
                test_stats_path = seed_dir / "test_metrics.json"
                if not test_stats_path.exists():
                    print(f"Missing {dataset} {model} {subject} {session} {seed}")
```
