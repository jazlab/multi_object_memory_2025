```python
"""Imports."""

from matplotlib import pyplot as plt
import numpy as np
```

```python
_PALETTE = {
    (0,): (1, 0, 0),
    (1,): (0, 1, 0),
    (2,): (0, 0, 1),
    (0, 1): (0.7, 0.7, 0),
    (0, 2): (0.7, 0, 0.7),
    (1, 2): (0, 0.7, 0.7),
}

fig, ax = plt.subplots(1, 1, figsize=(4, 4), dpi=200)
ax.set_aspect('equal')
ax.set_xlim(-1.35, 1.35)
ax.set_ylim(-1.35, 1.35)
ax.set_xticks([])
ax.set_yticks([])
cross = plt.Line2D([-0.05, 0.05], [0, 0], color='k')
ax.add_artist(cross)
cross = plt.Line2D([0, 0], [-0.05, 0.05], color='k')
ax.add_artist(cross)

# Make solid circle on ax at (1, 0)
for i in range(3):
    color = _PALETTE[(i,)]
    theta = i * 2 * np.pi / 3
    x = np.cos(theta)
    y = np.sin(theta)
    circle = plt.Circle((x, y), 0.18, color=color, fill=True)
    ax.add_artist(circle)

# Save figure
fig.savefig('colorscheme.pdf', bbox_inches='tight')
```
